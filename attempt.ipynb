```python
# open json file

import json
import pandas as pd

with open('massive_amr.jsonl') as f:
    data = [json.loads(line) for line in f]
    print(data[0])

for language in celtic_languages:
    translator = GoogleTranslator(source="en", target=celtic_languages[language])
    translated_sentences = []
    for i in range(10):
        text = data[i]['text']
        translated_text = translator.translate(text)
        translated_sentences.append(translated_text)
    # save to csv with columns: original, translated, language
    df = pd.DataFrame({'original': data[:10], 'translated': translated_sentences, 'language': language})
```

```python
import json
from transformers import MarianMTModel, MarianTokenizer

# ----------------------------
# Load Translation Model (English → Welsh)
# ----------------------------
model_name = "Helsinki-NLP/opus-mt-en-cy"  # English to Welsh
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

# ----------------------------
# Load JSONL Data
# ----------------------------
input_file = "massive_amr.jsonl"
output_file = "massive_amr_welsh.jsonl"

with open(input_file, "r", encoding="utf-8") as f:
    data = [json.loads(line) for line in f]

# ----------------------------
# Translate Function
# ----------------------------
def translate_text(text, tokenizer, model):
    """Translates English text to Welsh using MarianMT."""
    if not text.strip():
        return text  # Skip empty strings

    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=128)
    translated_tokens = model.generate(**inputs)
    translated_text = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]

    return translated_text

# ----------------------------
# Translate Sentences
# ----------------------------
for entry in data:
    entry["utt"] = translate_text(entry["utt"], tokenizer, model)
    entry["annot_utt"] = translate_text(entry["annot_utt"], tokenizer, model)
    print("done" + str(data.index(entry)) + "/" + str(len(data)))

# ----------------------------
# Save Translated Data
# ----------------------------
with open(output_file, "w", encoding="utf-8") as f:
    for entry in data:
        f.write(json.dumps(entry, ensure_ascii=False) + "\n")

print("✅ Translation complete! Saved as 'massive_amr_welsh.jsonl'")
```
